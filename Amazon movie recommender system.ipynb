```python
from surprise import Reader
from surprise import Dataset
from surprise import accuracy
from surprise import SVD
from surprise.model_selection import train_test_split
```

```python
movie_data = data.melt(id_vars = data.columns[0],value_vars=data.columns[1:],var_name="Movies",value_name="Rating")
movie_data
```

```python
rd = Reader(rating_scale=(-1,10))
newdata = Dataset.load_from_df(movie_data.fillna(0),reader=rd)
newdata
```

```python
train_data,test_data = train_test_split(newdata,test_size=0.20)
```

```python
svd = SVD()
```

```python
svd.fit(train_data)
```

```python
pred = svd.test(test_data)
```

```python
accuracy.rmse(pred)
```

```python
accuracy.mae(pred)
```
